```python
import cv2
import numpy as np
import tensorflow as tf

# Load your saved model
model = tf.keras.models.load_model("basic_2D_CNN.keras")

# Define class labels (replace with your actual class names)
class_names = {
    0: "Cover Drive",
    1: "Defense",
    2: "Flick",
    3: "Hook Shot",
    4: "Late Cut",
    5: "Lofted Shot",
    6: "Pull Shot",
    7: "Square Cut",
    8: "Straight Drive",
    9: "Sweep"
}

def preprocess_frame(frame):
    # Resize and normalize the frame
    resized = cv2.resize(frame, (224, 224))
    normalized = resized / 255.0
    return np.expand_dims(normalized, axis=0)

# Initialize webcam
cap = cv2.VideoCapture("/home/smayan/Downloads/All Cricket Shots Ever In Cricket History Part Two __ All Cricket Shots In Cricket __.mp4")

# Set a smaller display size for the window
display_width = 640
display_height = 900

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Preprocess the frame for prediction
    processed_frame = preprocess_frame(frame)
    
    # Make prediction
    predictions = model.predict(processed_frame)
    predicted_class = np.argmax(predictions[0])
    confidence = np.max(predictions[0])
    
    # Display prediction
    label = f"{class_names[predicted_class]}: {confidence:.2f}"
    cv2.putText(frame, label, (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Resize the frame for display
    small_frame = cv2.resize(frame, (display_width, display_height))
    
    # Show the smaller frame
    cv2.imshow('Cricket Shot Classification', small_frame)
    
    # Exit on 'q' press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
